# Visualizando Relações Estatísticas nos Dados.

```python
########################
# variáveis contínuas ##
########################

#replot e scatterplot
```

```python
##1.REPLOT
```

```python
#relação entre valor da gorjeta e da conta final
sns.relplot(x="total_bill", y="tip", data=tips)
```

```python
#Parâmetros semânticos: alteram a visualização dos dados em função do sentido
#exemplo: colocar cor ou tamanho para variáveis

#hue: adiciona cores a uma co-var
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

```python
#style: muda o símbolo de co-var
sns.relplot(x="total_bill", y="tip", hue="sex", style="sex", data=tips);
```

```python
#se a co-var for contínua, estabele uma paleta de cores (gradiente)
#documentacao das paletas de cores: https://seaborn.pydata.org/generated/seaborn.cubehelix_palette.html#seaborn.cubehelix_palette
sns.relplot(x="total_bill", y="tip", hue="size", data=tips);
```

```python
#size: muda o tamanho dos símbolos
sns.relplot(x="total_bill", y="tip", size="sex", data=tips)

#para alterar a escala de tamanho dos símbolos
sns.relplot(x="total_bill", y="tip", size="sex", sizes=(100,200), data=tips); #size(menor, maior)
```

```python
##1.SCATTERPLOT
sns.scatterplot(x="total_bill", y="tip", size="sex", data=tips);
```

```python
#########################
# variáveis categóricas #
#########################

#barplot, boxplot e scatterplot em grupos
```

```python
#barplot com barras na vertical: param x
sns.catplot(x="sex", kind="count", palette="Set2", data=tips);
```

```python
#barplot com barras na horizontal: param y
sns.catplot(y="day", kind="count", palette="Set1", data=tips);
```

```python
#scatterplot por grupo: 1 var categórica: x
sns.catplot(x="day", y="total_bill", palette="Set2", data=tips);
sns.catplot(x="day", y="total_bill", kind='swarm', data=tips); #sem sobreposição de pontos: param swarn
```

```python
#scatterplot por grupo: 2 var categórica: x e hue
sns.catplot(x="day", y="total_bill", kind='swarm', hue="sex", data=tips); #scatterplot com 2 var categóricas (x,hue)
```

```python
#alterar ordem das categorias em x
sns.catplot(x="day", y="total_bill", kind='swarm', hue="sex", order=['Sat','Sun','Fri','Thur'], data=tips);
```

```python
#boxplot: 1 var categórica: x
sns.catplot(x="day", y="total_bill", kind="box", data=tips);
```

```python
#boxplot: 2 var categórica x e hue
sns.catplot(x="day", y="total_bill", hue="smoker", kind="box", data=tips);
```

```python
#boxen: parecido com boxplot, mas dá peso para a largura das caixas
sns.catplot(x="day", y="total_bill", kind="boxen", data=tips);
```

```python
#gráficos combinados: boxplot + dispersão
sns.catplot(x="day", y="total_bill", kind="box", data=tips);
sns.stripplot(x="day", y="total_bill", data=tips, color="blue");
```

```python
#catplot em grid
#segregando pelos grupos da var time
sns.catplot(x="sex", y="total_bill", hue="smoker", col="time", data=tips, kind="box",height=4);
```

```python
# violin plot
sns.catplot(x="day", y="total_bill", kind="violin", data=tips) #vertical
sns.catplot(x="total_bill", y="day", kind="violin", data=tips) #horizontal
```

```python
#dividir o violin plot com o parâmetro split.
sns.catplot(x="day", y="total_bill", hue="sex", split=True , kind="violin", data=tips)
sns.catplot(x="day", y="total_bill", hue="sex", split=True , inner="stick", palette="pastel", kind="violin", data=tips);
```

```python
#fusão do violin plot com swarmplot
sns.catplot(x="day", y="total_bill", palette="pastel", kind="violin", data=tips);
sns.swarmplot(x="day", y="total_bill", color="k", size=3, data=tips)
```

```python
###############
## REGRESSÂO ##
###############

#lmplot
```

```python
# método lmplot: herda propriedades dos métodos scatterplot e lineplot
sns.lmplot(x="total_bill", y="tip", hue="sex", palette="Pastel1", data=tips);
```

```python
#dados por grupos: gráficos em cols e rows
#segregando pelos grupos da var time
sns.lmplot(x="total_bill",y="tip", hue="smoker", col="time", palette="Set2", data=tips);
```

## Cores no Seaborn

* A funcao que controla as cores no seaborn e a color_palette(). 
* Essa funcao fornece uma interface para gerar cores no seaborn.
* Usamos esse metodo internamente ao usar o parametro palette.
* O metodo color_palette() recebe o nome de uma paleta de cores do seaborn ou um colormap do matplotlib
* O metodo set_palette() recebe os mesmos parametros e eh usado para definir um estilo padrao para todos os plots.
* As cores devem ser definidas levando em consideracao o tipo de dado que queremos plotar. (qualitativos, sequenciais,diverging)


**Ferramenta Color Brewer 2.0 **

*  http://colorbrewer2.org/#type=sequential&scheme=BuGn&n=3

```python
#Visualzar paletas do color brew

# exemplo de paleta de cores para dados Quantitativos
sns.palplot(sns.color_palette("Accent"))

#dados divergetntes
sns.palplot(sns.color_palette("PuOr"))

#dados sequenciais (contínuos)
sns.palplot(sns.color_palette("Blues"))
```

```python
#Para plotar o gráfico com a paleta de cores do color brew basta copiar o nome da classe da paleta do site no param pallete do método

# gráfico de dispersão com paleta para dados contínuos
sns.relplot(x="total_bill", y="tip", hue="size", palette="Greens", size="size", sizes=(12,250), data=tips);
```

```python
# gráfico de barras com paleta para dados qualitativos
sns.catplot(x="sex", kind="count", palette="Set2", data=tips);
```

```python
# gráfico de heatmap com paleta para dados divergentes (paleta: param cmap - colour map)
sns.heatmap(tips.corr(), annot=True, cmap="PiYG")
```

```python
# Definindo cores padrão para gráficos Searborn
# todos os gráficos usarão essa paleta
sns.set_palette("pastel")

# plot usando a paleta de cor "pastel"
sns.relplot(x="total_bill", y="tip", data=tips);
```

```python
# Instrução with: plota o gráfico definindo uma opção de cor temporária != da padrão
with sns.color_palette("Accent"):
  sns.relplot(x="total_bill", y="tip", size="size", sizes=(12,250), data=tips);
```
